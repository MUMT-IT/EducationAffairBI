### Perform data summary using a group_by function

```python
def gen_pivot_table(df, groupby_cols, data_col, percentage=False):
    p = df.groupby(groupby_cols, as_index=False).count()
    if percentage:
        p['percent'] = p[data_col].apply(lambda x: x/p[data_col].sum()*100)
        cols = groupby_cols + [data_col] + ['percent']
    else:
        cols = groupby_cols + [data_col]
    d = p[cols]\
        .sort_values(groupby_cols).set_index(groupby_cols)
    d.to_clipboard()
    return d
```

```python
gen_pivot_table(df, ['Region', 'AddrProvinceDetail'], 'studentCode', percentage=True)
```

```python
gen_pivot_table(df, ['การทำงาน'], 'studentCode', percentage=True)
```

```python
d = gen_pivot_table(df, ['การทำงาน'], 'studentCode', percentage=True)
```
